```python
# === Shopping List Manager (Clean Version) ===
# This program lets the user add, view, remove, count, and exit.

# Variables to store up to 3 items
item1, item2, item3 = "", "", ""
count = 0

while True:
    # Menu
    print("\n=== Shopping List Manager ===")
    print("1. Add item")
    print("2. View items")
    print("3. Remove item")
    print("4. Count items")
    print("5. Exit")

    choice = input("Choose option: ")

    if choice == "1":  # Add item
        if item1 == "":
            item1 = input("Enter item: ")
            count += 1
            print("Item added!")
        elif item2 == "":
            item2 = input("Enter item: ")
            count += 1
            print("Item added!")
        elif item3 == "":
            item3 = input("Enter item: ")
            count += 1
            print("Item added!")
        else:
            print("Shopping list is full (max 3 items).")

    elif choice == "2":  # View items
        print("Current items:")
        if item1 != "":
            print("-", item1)
        if item2 != "":
            print("-", item2)
        if item3 != "":
            print("-", item3)
        if count == 0:
            print("No items in the list.")

    elif choice == "3":  # Remove item
        remove_item = input("Enter item to remove: ")
        if remove_item == item1:
            item1 = ""
            count -= 1
            print("Item removed.")
        elif remove_item == item2:
            item2 = ""
            count -= 1
            print("Item removed.")
        elif remove_item == item3:
            item3 = ""
            count -= 1
            print("Item removed.")
        else:
            print("Item not found in the list.")

    elif choice == "4":  # Count items
        print("Total items:", count)

    elif choice == "5":  # Exit
        print("Goodbye! Thank you for using Shopping List Manager.")
        break

    else:  # Invalid choice
        print("Invalid option. Please choose between 1â€“5.")
```
